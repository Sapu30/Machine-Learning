# 1. Import necessary libraries (matplotlib and seaborn). Load the csv file of dataset into dataframe in Python program. 

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("KCLT.csv")
df
```

# 2. Plot	min and max for actual_mean_temp, actual_min_temp and actual_max_temp using bar graph. Use blue for min and red for max.

```python
n_groups = 3
plt. rcParams["figure.figsize"] = (7,7)
means_min = (min(df["actual_mean_temp"]),min(df['actual_min_temp']),min(df['actual_max_temp']))
means_max = (max(df["actual_mean_temp"]),max(df['actual_min_temp']),max(df['actual_max_temp']))


# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.32
bar_width2 = 0.16
opacity = 1.0

rects1 = plt.bar(index, means_min, bar_width,
alpha=opacity,
color='b',
label='Minimum')

rects2 = plt.bar(index + bar_width, means_max, bar_width,
alpha=opacity,
color='r',
label='Maximum')

plt.xlabel('\n Attributes')
plt.ylabel('Temperature')
plt.title('Minimum and Maximum values of temperature \n')
plt.xticks(index + bar_width2,('actual_mean_temp', 'actual_min_temp','actual_max_temp'))
plt.legend()

plt.tight_layout()
plt.show()
```

# 3. Plot	two separate histograms for actual_mean_temp for months of Aug 2014 and March 2015. Make key observations (put them as comments in jupyter notebook).

```python
#Histogram for Actual Mean Temperature of August 2014

date1 = ((df["date"]>="2014-8-1") & (df["date"]<="2014-8-31"))
filter_dates1 = df.loc[date1]
li1 = list(filter_dates1["actual_mean_temp"])
fig1 = plt.hist(li1, facecolor="green")
plt.xlabel('\n Attributes')
plt.ylabel('Days')
plt.title('Histogram for Actual Mean Temperature of August 2014 \n')
plt.show()
#observation for August 2014 is that the Actual mean temperature range is from 68 to above 82. 77 to 78 has been the temperature 
#for maximum days. Also there is no record of temperature 76 in August 2014 as the plot shows.


#Histogram for Actual Mean Temperature of March 2015
import matplotlib.pyplot as plt
date2 = ((df["date"]>="2015-3-1") & (df["date"]<="2015-3-31"))
filter_dates2 = df.loc[date2]
li2 = list(filter_dates2["actual_mean_temp"])
fig2 = plt.hist(li2, facecolor="orange")   
plt.xlabel('\n Attributes')
plt.ylabel('Days')
plt.title('Histogram for Actual Mean Temperature of March 2015 \n')
plt.show()
# observation for March 2015 is that the range of Actual mean temperature has been a little above 35 to a little above 70. Also 
# temperature between 50 to 55 and 60 to 65 has been recorded for maximum number of days.
```

# 4. Plot box plots for attributes actual_mean_temp, actual_min_temp and actual_max_temp. Make key observations (put them as comments in jupyter notebook).

```python
plt. rcParams["figure.figsize"] = (7, 7)
df.boxplot(column=["actual_mean_temp", "actual_min_temp", "actual_max_temp"]) 
plt.xlabel("\n Attributes")
plt.ylabel("Temperature")
plt.title("Boxplot of Temperature\n")

# observationis that the maximum value for actual_mean_temp lies between 80 to 100, for actual_min_temp it lies between 0 to 100
# for actual_max_temp it is exactly 100. The minimum value for actual_mean_temp lies a little below 20, for actual_min_temp
# it lies between 0 to 20, closer to 0 which indicated being below 10, and for actual_max_temp, it lies between 20 to 30.
# Also for the column actual_mean_temp, the enteries after the 1st quartile mark to median is more than from median to 3rd 
# quartile mark.
```

# 5. Find the correlation between the following pairs of attributes: actual_mean_temp & actual_min_temp and actual_mean_temp & actual_max_temp. Make key observations. 

```python
#Correlation between the actual_mean_temp & actual_min_temp.
plt. rcParams["figure.figsize"] = (7, 7)
df1 = df[["actual_mean_temp","actual_min_temp"]]
corrMatrix1 = df1.corr()
sns.heatmap(corrMatrix1, annot = True)
plt.show()
# observation for first pair(according to the question): the attributes are hightly postively correlated, with a correlation of 0.97 
# which means they are closely related for different attributes. For the same attribute pair(i.e 2 same attribures on x and y 
# axis) are perfectly positive correlated with a correlation of 1


#Correlation between the actual_mean_temp & actual_max_temp.
plt. rcParams["figure.figsize"] = (7, 7)
df2 = df[["actual_mean_temp","actual_max_temp"]]
corrMatrix2 = df2.corr()
sns.heatmap(corrMatrix2, annot = True,cmap = "mako")
plt.show()
# observation for first pair(according to the question): the attributes are hightly postively correlated, with a correlation of 0.97 
# which means they are closely related for different attributes. For the same attribute pair(i.e 2 same attribures on x and y 
# axis) are perfectly positive correlated with a correlation of 1
```

# 6. Plot	two scatter plots between the following pairs of attributes: 	actual_mean_temp & actual_min_temp and actual_mean_temp & actual_max_temp. Make key observations.

```python
#Scatter plots between the actual_mean_temp & actual_min_temp.
plt. rcParams["figure.figsize"] = (7, 7)
l1 = df["actual_mean_temp"]
l2 = df["actual_min_temp"]
plt.xlabel("\n Actual Mean Temperature")
plt.ylabel("Actual Minimum Temperature")
plt.title("\nScatter plot between Actual Mean Temperature and Actual Minimum Temperature\n")
plt.scatter(l1,l2, color = "orange")
plt.show()
# observation - the density of dots is more in near 75 to 85 range in the x axis i.e the two attributes have more higher 
# valuesthen the lower values i.e Mostly the temperature has been ho the higher side in mean temperature in range 75 to 85
# and minimum temperature in 60 to 70 for maximum days.


#Scatter plots between the actual_mean_temp & actual_max_temp.
plt. rcParams["figure.figsize"] = (7, 7)
l1 = df["actual_mean_temp"]
l3 = df["actual_max_temp"]
plt.xlabel("\n Actual Mean Temperature")
plt.ylabel("Actual Maximum Temperature")
plt.title("\nScatter plot between Actual Mean Temperature and Actual Maximum Temperature\n")
plt.scatter(l1,l3, color="darkblue")
plt.plot()
plt.show()
# observation - the density of dots is more in 70 to 80 range in the x axis i.e the two attributes have more higher values
# then the lower values i.e Mostly the temperature has been ho the higher side in mean temperature in range 70 to 80
# and maximum temperature in 80 to 90 for maximum days.
```
