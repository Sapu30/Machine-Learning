<h2>Task : 4
<h2>TEXT DATA:

```python
data=pd.read_csv("a3-Q4.csv")
data
```

```python
#drop unwanted attribute
data.drop(data.columns[[0,1,2]], axis = 1,inplace = True)
data
```

<h2>4.1 Split into Words

```python

data['split_content'] = data['content'].str.split()
```

```python
data
```

<h2>4.2 Filter out Punctuation

```python
import re
import string
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation,'')
    return text
```

```python
data['clean_content'] = data['content'].apply(remove_punctuations)
data
```

<h2>4.3 Filter out stop words

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stopwords = stopwords.words('english')
```

```python
def remove_stopwords(txt_tokenized):
    txt_clean = [word for word in txt_tokenized if word not in stopwords]
    return txt_clean

data['stop_content'] = data['split_content'].apply(lambda x: remove_stopwords(x))
data
```

<h2>4.4 Stem words

```python

from nltk.stem import PorterStemmer
ps = PorterStemmer()
dir(ps)
```

```python
pd.set_option('display.max_colwidth',100)
```

```python
def stemming(tokenized_text):
    text = [ps.stem(word) for word in tokenized_text]
    return text

data['stemmed'] = data['stop_content'].apply(lambda x: stemming(x))
data
```
